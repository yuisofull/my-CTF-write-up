```sage
from Crypto.Util.number import getPrime, bytes_to_long
import random
import os

e = 65537
p = getPrime(1024)
while (p - 1) % e == 0:
    p = getPrime(1024)

Ns = []
while len(Ns) < 4:
	q = getPrime(2048)
	N = p * q
	if (q - 1) % e != 0 and N not in Ns:
		Ns.append(N)

# size N = 3072 bits

flag = open("flag.txt", "rb").read()
flag += os.urandom(3072 // 8 - 2 - len(flag)) # 382-len(flag) bytes

flag = bytes_to_long(flag)

cts = [pow(flag, e, N) for N in Ns] # flag^e mod N = ct
Ns = [N + int(os.urandom(32).hex(), 16) for N in Ns] # 256 bits of randomness

# p*q1 + a= N1=>N1 >> 256 = p*q1>>256
# p*q2 + b= N2
# p*q3 + c= N3
# p*q4 + d= N4


random.shuffle(Ns)

f = open("output.txt", "w")
f.write(f"{Ns = }\n")
f.write(f"{cts = }\n")
f.close()
```

```sage
from sage.all import *
lam = 256
# 4x4 matrix
B = [[0 for _ in range(4)] for _ in range(4)]
B[0][0] = 1<<(lam+1)
for i in range(1,4):
    B[0][i] = Ns[i]
    B[i][i] = -Ns[0]
B=matrix(ZZ,B)
# print(B)
B = B.LLL()
```

```sage
# decrypt
from Crypto.Util.number import long_to_bytes
e= 65537
def decrypt(ct, p, q):
    n = p*q
    phi = (p-1)*(q-1)
    d = pow(e, -1, phi)
    return pow(ct, d, n)
print(long_to_bytes(decrypt(cts[0], p, q0)))
print(long_to_bytes(decrypt(cts[1], p, q1)))
print(long_to_bytes(decrypt(cts[2], p, q2)))
print(long_to_bytes(decrypt(cts[3], p, q3)))
```
